## Mosquito Species

Transmission of WNV is most commonly spread through the bite of an infected mosquito, where the mosquitoes themselves are infected from feeding on infected birds ([*source*](https://www.cdc.gov/westnile/transmission/index.html)). Birds act as reservoirs of the virus, with the virus multiplying in their blood for several days ([*source*](https://www.who.int/news-room/fact-sheets/detail/west-nile-virus)). 

Culex mosquitoes are the main vectors of the virus, especially Culex pipiens ([*source*](https://www.who.int/news-room/fact-sheets/detail/west-nile-virus)). 

```python
plt.style.use('seaborn-poster')
sns.set_style(style = 'white')

# Visualise number of traps by WNV
plt.figure(figsize=(10,6))
sns.countplot(data=train, x='wnv_present')
plt.title('Number of traps split by WNV presenece', fontsize=15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);
```

From the count plot above, we see that there is a imbalance in class. There is a significantly higher number of traps with WNV absent compared to traps with WNV present. We will have to take this into consideration during modelling.

```python
# Visualisation of mosquito species against number of mosquito and presence of WNV 
mos = train[['species', 'num_mosquitos', 'wnv_present']]

mos_wnv = mos.groupby(by=['species']).sum()
print(mos_wnv)

# Plot graph
plt.figure(figsize=(10,6))
mos.groupby('species')['num_mosquitos'].sum().plot(kind='barh')
mos.groupby('species')['wnv_present'].sum().plot(kind='barh', color='orangered')
plt.title('Number of mosquitoes and presence of WNV per species', fontsize=15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);
```

```python
# Since number of mosquitoes with WNVpresent is too small to visualise against number of mosquitoes,
# Here's a closer look at WNV present graph
plt.figure(figsize=(10,6))
mos.groupby('species')['wnv_present'].sum().plot(kind='barh', color='orangered')
plt.title('Number of mosquitoes with WNV present', fontsize=15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);
```

From above, we see that Culex Restuans and Culex Pipens are the most captured, as well as the only mosquitoes that carries the West Nile Virus in this area. Out of the total mosquitoes sampled, only 0.4% tested positive with WNV.


## Spray Data

It appears that spraying occurred in two main blocks. The first time was on two dates in August and September of 2011. The second time was a more prolonged spraying campaign, taking place over 8 dates in July, August, and September of 2013. The second spray campaign was also more intensive in terms of the number of locations sprayed. As seen on the left, the peak of the campaign was in August 2013. Across the 5 dates in August 2013, spraying took place in nearly 8000 locations.

This was likely due to a large spike in infected mosquitos during that period. As seen on the right, there was a large spike in cases in August and September 2013. At the peak of the 2013 outbreak, over 25% percent of traps contained infected mosquitos.

```python
# Plot number of locations sprayed by date
plt.figure(figsize = (15,5))
plt.subplot(1, 2, 1)
spray.groupby(["date"]).size().plot(kind = "bar", title = "Number of locations sprayed");
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

# Plot percentage of traps containing mosquitos with WNV
plt.subplot(1,2,2)
train.groupby(["date"])["wnv_present"].mean().plot(kind = "line", title = "Percentage of traps containing mosquitos with WNV")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);
```

However, a closer look at the year 2013 reveals that spraying did not necessarily reduce the number of mosquitos carrying WNV. The red dotted lines represent the dates on which spraying took place. Spraying wasn't strongly associated with decrease in percentage of traps containing mosquitos with WNV.

As mosquito sprays are generally effective up to 300 feet from the spray location ([source](https://www.montcopa.org/DocumentCenter/View/8932/FAQMosquitoSpraying?bidId=)), we used geopy to ascertain which trap were within effective range of spray location. From the table below, we can see that out of 168 traps in which WNV was detected, only 32 of them (or 19% of them) were within effective range of at least one spray location. This indicates that spray locations were not effectively positioned at areas where WNV was detected.

```python
# Defining function to record how many spray locations were within 300 feet of the trap
def near_spray(cell):
    nearby_locs = 0
    for spray_coord in spray["coordinates"]:
        if geopy.distance.distance(cell, spray_coord).feet < 300:
            nearby_locs += 1
    return nearby_locs
```

## Time and Seasonality

Let's take a look at WNV over time and different seasons. 

The mosquito population is the highest in summer, and the WNV season generally reaches its height in  August and September ([*source*](https://www.webmd.com/a-to-z-guides/west-nile-virus-faq)). The mosquito season in Illinois is from early April to mid-October ([*source*](https://cristtermite.com/2021/10/when-does-mosquito-season-end-in-illinois/)) 

```python
# Look at WNV over the years
plt.figure(figsize=(10,6))
train.groupby('year')['wnv_present'].sum().plot(kind='bar')
plt.xticks(rotation=45)
plt.xlabel('Year', fontsize=15)
plt.title('WNV Present - Years', fontsize=15);
```

There is no clear trend between WNV and year. WNV looks higher in 2007 and 2013 (with over 200 detections), while 2009 and 2011 have less than 60 detections.

```python
# Look at WNV over the months
plt.figure(figsize=(10,6))
train.groupby('month')['wnv_present'].sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel('Month', fontsize=15)
plt.title('WNV Present - Months (2007, 2009, 2011, 2013)', fontsize=15);
```

WNV seems to peak in August, with the beginning of WNV starting from June and ending in October.

```python
# Look at WNV over the weeks
plt.figure(figsize=(10,6))
train.groupby('week')['wnv_present'].sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel('Week', fontsize=15)
plt.title('WNV Present - Weeks (2007, 2009, 2011, 2013)', fontsize=15);
```

Similar to our observation in the month plot, we find the peak to be in week 34 which corresponds to mid to end-August. The beginning of WNV season starts from week 26 (end-June) and ends in week 40 (start of October).

Looking at both the plots for month and week, our findings are similar to reported trends of WNV, with the peak of WNV coninciding with summer in Chicago and stretching into autumn (October/November). A possible explanation for this could be the warmer months in summer and early-autumn that encourages mosquito propagation.

```python
# Plot sampling efforts by year
plt.figure(figsize=(10,6))
train['year'].value_counts().plot(kind ='barh')
plt.title('Sampling efforts by year', fontsize=15);
```

```python
# Plot WNV presence against month, year
plt.figure(figsize=(10,6))
mth_year = pd.pivot_table(train,index=['month'],columns=['year'], values = 'wnv_present')

sns.lineplot(data=mth_year)
plt.title('Monthly WNV Presence by Year', fontsize=15);
```

The bar graph showing total number of traps placed seems to be highest in 2007, with traps staying rather constant for subsequent years. The line plt shows the spread of WNV detected monthly by year. There is a clear peak in August each year which agrees with our findings earlier from our monthly and weekly plots. 2013 also has the highest number of detections.

```python
# Now that we've seen WNV trends, let's see if the trends for number of mosquitoes follow a similar trend

fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize=(18,15))

# Look at number of mosquitoes over the years
train.groupby('year')['num_mosquitos'].sum().plot(kind='bar', ax=ax[0,0])
ax[0,0].set_xlabel('Year', fontsize=15)
ax[0,0].set_title('Number of Mosquitoes - Years', fontsize=15)
ax[0,0].tick_params(labelrotation=0);

# Look at number of mosquitoes over the months
train.groupby('month')['num_mosquitos'].sum().plot(kind='bar', ax=ax[0,1])
ax[0,1].set_xlabel('Month', fontsize=15)
ax[0,1].set_title('Number of Mosquitoes - Months (2007, 2009, 2011, 2013)', fontsize=15)
ax[0,1].tick_params(labelrotation=0);

# Look at number of mosquitoes over the weeks
train.groupby('week')['num_mosquitos'].sum().plot(kind='bar', ax=ax[1,0])
ax[1,0].set_xlabel('Week', fontsize=15)
ax[1,0].set_title('Number of Mosquitoes - Weeks (2007, 2009, 2011, 2013)', fontsize=15)
ax[1,0].tick_params(labelrotation=0);

# Plot number of mosquitoesagainst month, year
mth_year = pd.pivot_table(train,index=['month'],columns=['year'], values = 'wnv_present')
sns.lineplot(data=mth_year, ax=ax[1,1])
ax[1,1].set_title('Monthly Number of Mosquitoes by Year', fontsize=15);
```

The trends in number of mosquitoes follows that of WNV, which is expected as WNV presence in mosquitoes is a subset of the total number of mosquitoes during that period.

From preliminary visualisations, we can see spikes in the month of August in 2007 and 2013. As such, we would like to dig deeper to see if there is any relationship between the presence of West Nile Virus with regards to seasons.

The season will be coded as follows ([source](https://www.climatestotravel.com/climate/united-states/chicago)):
* Winter: December to February 
* Spring: March to May 
* Summer: June to August 
* Autumn: September to November 

```python
train['season'] = train['month'].map({
        12: 'summer', 1: 'summer', 2: 'summer',
        3: 'spring', 4: 'spring', 5: 'spring',
        6: 'summer', 7: 'summer', 8: 'summer',
        9: 'autumn', 10: 'autumn', 11:'autumn'
    })
```

```python
train.groupby(['season'])['wnv_present'].sum().plot(kind='bar')
plt.title('Effect of seasonality on WNV')
plt.xticks(rotation = 0);
```

As expected, we see a higher number of WNV cases during the summer months as compared to the other seasons. This could be due to the temperature and humidity which will be explored later in the weather EDA.

### Humidity

At 100% relative humidity, the wet-bulb temperature is equal to the air temperature (`Tavg`); at lower humidity the wet-bulb temperature is lower than dry-bulb temperature because of evaporative cooling ([source](https://en.wikipedia.org/wiki/Wet-bulb_temperature)). As the total precipation, do not have high correlation number of mosquitos, we will explore the humidity by getting the difference betweem `tavg` and `wet_bulb`.

```python
# Check if there is any Wet Bulb temperature above Tavg.
train[train['wet_bulb'] >= train['tavg']]
```

As we do not have any `wet_bulb` temperature greater than or equals to `tavg`, there is no 100% relative humidity.

Hence, the greater the difference between `wet_bulb` and `tavg`, the lower the relative humidity.

```python
train['humidity'] = train['tavg'] - train['wet_bulb']
train[['humidity','wet_bulb','tavg']].head()
```

```python
# Check Humidity column
train['humidity'].describe()
```

```python
# Plot West Nile Virus vs Relative Humidity
train.groupby(['humidity'])['wnv_present'].sum().plot.bar(color='#8383EC') 
plt.title(f"West Nile Virus vs Relative Humidity", fontsize=16)
plt.xlabel("Relative Humidity", fontsize =13)
plt.xticks(rotation=90)
plt.ylabel("Sum of West Nile Virus detections", fontsize=13);
```

```python
# Plot Total Number of Mosquitos vs Relative Humidity
train.groupby(['humidity'])['num_mosquitos'].sum().plot.bar(color='#8383EC') 
plt.title(f"Total Number of Mosquitos vs Relative Humidity", fontsize=16)
plt.xlabel("Relative Humidity", fontsize =13)
plt.xticks(rotation=90)
plt.ylabel("Total Number of Mosquitos", fontsize=13);
```

With reference to this [source](https://www.livescience.com/40663-measuring-relative-humidity-science-fair-projects.html) and using the mean `tavg` value, and the difference of `tavg` and `wet bulb` (differences of 6 to 10 is the higher few) we found out that the optimum temperature for WNV positive mosquitoes is about 39 to 66% relative humidity, with 66% being the most ideal.

We also found out that the optimum temperature for number of mosquitos is about 33 to 66% relative humidity, with 33% being the most ideal. 

However **this does not align with the [source](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3750695/) we have found**, indicating that the **ideal humidity level of mosquitos are around 50 to 90% relative humidity**. Mosquitoes also have a 40% shorter lifespan if humidity levels were to fall below 40% if they are held without food or water.

This shows that the rough estimate using `tavg` and `wet_bulb` is not accurate and we will have to feature engineer a specific column to calculate humidity in the later part.
